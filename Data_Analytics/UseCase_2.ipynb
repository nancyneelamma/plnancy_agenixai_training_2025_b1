```python
# Copying the dataset into another variable, to use and make changes, by not affecting the original dataset

stud_data=df.copy()
print(stud_data)
```

```python
# Display entire information of the dataframe

stud_data.info()
```

```python
# Classify the ages into Adolescents,Young Adults

def age_cat(age):
  if(age >= 15 and age <=18 ):
    return "Adolescents"
  elif(age >= 19):
    return "Young Adults"
  

# Add them to the dataframe

stud_data['Age_Category'] = stud_data['Age'].apply(age_cat)
```

```python
# Convert attribute sex from numerical values into categorical values

def gender(sex):
    if sex == 'Female':
      return 0
    elif sex == 'Male':
      return 1

# Add them to the dataframe

stud_data['Gender_Category'] = stud_data['Gender'].apply(gender)
```

```python
# Convert attribute target from numerical values into categorical values

def socioeconomic_status(socioeconomic):
    if socioeconomic == 'Low':
        return 0
    elif socioeconomic == 'Middle':
        return 1
    elif socioeconomic == 'High':
        return 2

# Add them to the dataframe

stud_data['Socio_Economic_Category'] = stud_data['SocioeconomicStatus'].apply(socioeconomic_status)
```

```python
# Convert attribute target from numerical values into categorical values

def academicperformance_status(academicperformance):
    if academicperformance == 'Low':
        return 0
    elif academicperformance == 'Middle':
        return 1
    elif academicperformance == 'High':
        return 2

# Add them to the dataframe

stud_data['Academic_Performance_Category'] = stud_data['AcademicPerformanceStatus'].apply(academicperformance_status)
```

```python
# Convert attribute target from numerical values into categorical values

def classParticipation(classparticipation):
    if classparticipation == 'Low':
        return 0
    elif classparticipation == 'Medium':
        return 1
    elif classparticipation == 'High':
        return 2

# Add them to the dataframe

stud_data['Class_Participation_Category'] = stud_data['ClassParticipation'].apply(classParticipation)
```

```python
# Convert numerical/categorical attributes into meaningful labels

def academicPerformance_status(aps):
    if aps == 'Fail':
        return 0
    elif aps == 'Pass':
        return 1
    
# Add them to the dataframe

stud_data['Academic_Performance_Category'] = stud_data['AcademicPerformanceStatus'].apply(academicPerformance_status)
```

```python
# Descriptive Statistics

stud_data.describe()
```

```python
# Drop old numerical/categorical columns if no longer needed

stud_data.drop(columns=['StudentID'], inplace=True)
```

```python
# Taking look at the data by Top 10 rows

stud_data.head(10)
```

```python
# Finding correlation between attributes using heatmap

correlation_matrix = stud_data.corr(numeric_only=True)

#Plotting

plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Attributes")
plt.show()
```

```python
# Grades vs. Academic Performance

plt.figure(figsize=(8, 5))
sns.boxplot(x='AcademicPerformanceStatus', y='Grades', hue='AcademicPerformanceStatus', data=stud_data, palette='Set3')
plt.title('Grades vs. Academic Performance')
plt.xlabel('Academic Performance')
plt.ylabel('Grades')
plt.show()
```

```python
#  Age vs Time Spent on homework   

plt.figure(figsize=(8, 5))
sns.lineplot(x='Age', y='TimeSpentOnHomework', data=stud_data, marker='o', color='purple')
plt.title('Age vs. Time Spent on Homework')
plt.xlabel('Age')
plt.ylabel('Time Spent on Homework (Hours)')
plt.show()
```

```python
#  Academic_Performance_Category vs Time Spent on homework  

plt.figure(figsize=(8, 5))
sns.boxplot(x='AcademicPerformanceStatus', y='TimeSpentOnHomework', hue='AcademicPerformanceStatus',data=stud_data, palette='Set1')
plt.title('Academic Performance Category vs. Time Spent on Homework')
plt.xlabel('Academic Performance')
plt.ylabel('Time Spent on Homework (Hours)')
plt.show()
```

```python
# ClassParticipation vs Attendance

plt.figure(figsize=(8, 5))
sns.countplot(x='ClassParticipation', data=stud_data, hue='ClassParticipation',palette='coolwarm')
plt.title('Class Participation vs. Attendance')
plt.xlabel('Class Participation Level')
plt.ylabel('Count')
plt.show()
```

# Summary

The goal is to understand how factors like attendance, homework, class participation, and family background impact academic performance. By studying this, schools can take steps to help students perform better and score better grades.

# Methodology

a) The students information was cleaned by checking for null and duplicates, dropped unecessary columns and converted categorical values into numerical values using label encoding so it could be analyzed properly.

b) Calculated Mean, Median and Standard Deviation Using Descriptive Statistics.

c) To find the co-relation between the attributes used heatmap.

d) Performed Exploratory Data Analysis where visualizations were generated to analyze study the connection between grades, attendance, homework, and class participation.


# Key Findings

## Grades & Academic Performance

Students with higher grades are more likely to pass.

## Age vs Time Spent on homework  

Students above the age of 20 years old spend maximum time on home works and below age of 18 years old tend to spend slightly more time on homework compared to students of age 18 and 19 years old, possibly due to stricter academic expectations.


## Academic Performance vs Time Spent on homework 

Students who pass generally spend more time on homework compared to failing students.

## Class Participation & Attendance

Students who participate more in class tend to attend more regularly leads to better engagement and learning and participations is average.


##  Age & Grades

Younger students have a wider range of grades, while older students show more stable performance.
