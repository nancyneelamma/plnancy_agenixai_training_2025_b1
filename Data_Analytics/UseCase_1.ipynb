```python
# Copying the dataset into another variable, to use and make changes, by not affecting the original dataset

cc_data=df.copy()
print(cc_data)
```

```python
# Check for duplicates

duplicates = cc_data.duplicated()
number_duplicates = duplicates.sum()
print(number_duplicates)
```

```python
# Display entire information of the dataframe

cc_data.info()
```

```python
# Classify the ages into Young Adults, Mid-Age Adults, Old Age

def age_cat(age):
  if(age >= 18 and age <= 29):
    return "Young Adults"
  elif(age >= 30 and age <= 59):
    return "Mid-Age Adults"
  elif(age >= 60):
    return "Old Age"

# Add them to the dataframe

cc_data['Age_Category'] = cc_data['Age'].apply(age_cat)
```

```python
# Convert attribute target from numerical values into categorical values

def map_location(location):
    if location == 'Urban':
        return 0
    elif location == 'Rural':
        return 1
    elif location == 'Suburban':
        return 2

# Add them to the dataframe

cc_data['Location_Category'] = cc_data['Location'].apply(map_location)
```

```python
# Convert numerical/categorical attributes into meaningful labels

def plan_type(plan):
    if plan == 'Basic':
        return 0
    elif plan == 'Premium':
        return 1
    
# Add them to the dataframe

cc_data['Plan_Category'] = cc_data['PlanType'].apply(plan_type)
```

```python
# Convert numerical/categorical attributes into meaningful labels

def payment_method(payment):
    if payment == 'CreditCard':
        return 0
    elif payment == 'BankTransfer':
        return 1
    
# Add them to the dataframe

cc_data['Payment_Category'] = cc_data['PaymentMethod'].apply(payment_method)
```

```python
# Descriptive Statistics

cc_data.describe()
```

```python
# Convert numerical/categorical attributes into meaningful labels

def churn_status(churn):
    if churn == 'No':
        return 0
    elif churn == 'Yes':
        return 1
    
# Add them to the dataframe

cc_data['Churn_Category'] = cc_data['ChurnStatus'].apply(churn_status)
```

```python
# Drop old numerical/categorical columns if no longer needed
cc_data.drop(columns=['CustomerID'], inplace=True)
```

```python
# Taking look at the data by Top 10 rows

cc_data.head(10)
```

```python
# Finding correlation between attributes using heatmap

correlation_matrix = cc_data.corr(numeric_only=True)

#Plotting

plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Attributes")
plt.show()
```

```python
# Churn Status by Age Distribution  

plt.figure(figsize=(12,6))
sns.countplot(x='Age_Category', hue='ChurnStatus', data=cc_data, palette='viridis')
plt.title("Churn by Age Distribution ")
plt.xlabel("Age Category")
plt.ylabel("Count")
plt.show()
```

```python
#  Churn Status by Location 

plt.figure(figsize=(8, 5))
sns.countplot(x='Location', hue='ChurnStatus', data=cc_data)
plt.title('Churn by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```

```python
# Churn by Plan Type

plt.figure(figsize=(8, 5))
sns.countplot(x='PlanType', hue='ChurnStatus', data=cc_data, palette='coolwarm')
plt.title('Churn by Plan Type')
plt.xlabel('Plan Type')
plt.ylabel('Count')
plt.show()
```

```python
#  Churn by Payment Method 

plt.figure(figsize=(8, 5))
sns.countplot(x='PaymentMethod', hue='ChurnStatus', data=cc_data)
plt.title('Payment Method vs. Churn')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
# Data Usage vs. Plan Category

plt.figure(figsize=(8, 5))
sns.boxplot(x='Plan_Category', y='DataUsage', hue='Plan_Category', data=cc_data, palette='muted')
plt.title('Data Usage vs. Plan Category')
plt.xlabel('Plan Category')
plt.ylabel('Data Usage (GB)')
plt.xticks(ticks=[0, 1], labels=['Basic', 'Premium'])
plt.show()
```

```python
# Payment Category vs. Location

plt.figure(figsize=(8, 5))
sns.countplot(x='Payment_Category', hue='Location', data=cc_data, palette='coolwarm')
plt.title('Payment Category vs. Location')
plt.xlabel('Payment Category')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['CreditCard', 'BankTransfer'])
plt.show()
```

# REPORT


# Summary

The goal is to find the reasons behind customer churn and suggest ways to retain customers. The study examines customer details, service usage, and billing habits to determin what makes people stay or leave.


# Methodology

a) The customer information was cleaned by checking for null and duplicates, dropped unecessary columns and converted categorical values into numerical values using label encoding so it could be analyzed properly.

b) Calculated Mean, Median and Standard Deviation Using Descriptive Statistics.

c) To find the co-relation between the attributes used heatmap.

d) Performed Exploratory Data Analysis where visualizations were generated to analyze customer demographics, usage patterns, billing information, and churn trends.


# Key Findings

## Age Category and Churn Status

a) Mid-age adults (30-59) represent the majority of the customer base.

b) Younger and older customers show slightly higher churn rates, possibly due to pricing.

## Churn by Location

a) Customers in Urban and Rural areas exhibit higher churn rates compared to Suburban areas.

b) Service quality in urban areas might be influencing this trend.

## Plan Type and Churn

a) Basic plan users churn at a higher rate than Premium plan users.

b) Premium customers are likely receiving better service or more suitable plans.

## Payment Method and Churn

a) Customers using Credit Cards tend to churn more than those using Bank Transfers.

b) Bank Transfers may lead to increased retention.

## Data Usage and Plan Type

a) Premium plan users tend to use more data than Basic plan users.

b) Data-heavy users might be less likely to churn if their needs are met.

## Payment Category and Location

a) The data suggests that the company's customer base is concentrated in urban and rural areas.

b) Credit card payment is the most popular payment method.


# Recommendations

a) Offer discounts or customized plans for younger and older customers who are more likely to churn.

b) Evaluate competitor offerings, especially in urban areas, and adjust pricing and service quality accordingly.

c) Consider revising pricing for Basic plan users to increase retention.

d) Introduce flexible payment options for credit card users to improve retention.

e) Conduct satisfaction surveys to understand the reasons behind churn.

f) Conduct targeted promotions in urban areas where churn rates are higher.

g) Enhance customer service support, particularly for high-paying and high-usage customers.
