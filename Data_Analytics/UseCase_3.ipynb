```python
# Copying the dataset into another variable, to use and make changes, by not affecting the original dataset

energy_data=df.copy()
print(energy_data)
```

```python
# Taking look at the data by Top 10 rows

energy_data.head(21)
```

```python
# Check null values

energy_data.isnull().sum()
```

```python
# Filling Null Values in Renewable Energy Source using 'None'

energy_data.loc[:, 'RenewableEnergySource']=energy_data['RenewableEnergySource'].fillna('None')
```

```python
# Check null values

energy_data.isnull().sum()
```

```python
# Taking look at the data by Top 10 rows

energy_data.head(21)
```

```python
# Check for duplicates

duplicates = energy_data.duplicated()
number_duplicates = duplicates.sum()
print(number_duplicates)
```

```python
# Display entire information of the dataframe

energy_data.info()
```

```python
# Convert attribute target from numerical values into categorical values

def insulationTypes(insulationtype):
    if insulationtype == 'Poor':
        return 0
    elif insulationtype == 'Fair':
        return 1
    elif insulationtype == 'Good':
        return 2
    elif insulationtype == 'Excellent':
        return 3

# Add them to the dataframe

energy_data['InsulationType_Category'] = energy_data['InsulationType'].apply(insulationTypes)
```

```python
# Convert attribute target from numerical values into categorical values

def renewableEnergySource(renewableenergysource):
    if renewableenergysource == 'None':
        return 0
    elif renewableenergysource == 'Solar':
        return 1
    elif renewableenergysource == 'Wind':
        return 2

# Add them to the dataframe

energy_data['Renewable_EnergySource_Category'] = energy_data['RenewableEnergySource'].apply(renewableEnergySource)
```

```python
# Descriptive Statistics

energy_data.describe()
```

```python
# Drop old numerical/categorical columns if no longer needed

energy_data.drop(columns=['BuildingID'], inplace=True)
```

```python
# Taking look at the data by Top 10 rows

energy_data.head(10)
```

```python
# Finding correlation between attributes using heatmap

correlation_matrix = energy_data.corr(numeric_only=True)

#Plotting

plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Attributes")
plt.show()
```

```python
# Energy Consumption vs Building size

plt.figure(figsize=(8, 5))
sns.scatterplot(x='BuildingSize', y='EnergyConsumption', data=energy_data, alpha=0.6, color='blue')
plt.title('Energy Consumption vs. Building Size')
plt.xlabel('Building Size (sq ft)')
plt.ylabel('Energy Consumption (kWh)')
plt.show()
```

```python
# Humidity vs Insulation Type

plt.figure(figsize=(8, 5))
sns.boxplot(x='InsulationType', y='Humidity', data=energy_data, hue='InsulationType', palette='coolwarm', legend=False)
plt.title('Humidity vs. Insulation Type')
plt.xlabel('Insulation Type')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
# Building Size vs Insulation Type

plt.figure(figsize=(8, 5))
sns.boxplot(x='InsulationType', y='BuildingSize', data=energy_data, hue='InsulationType', palette='muted', legend=False)
plt.title('Building Size vs. Insulation Type')
plt.xlabel('Insulation Type')
plt.ylabel('Building Size (sq ft)')
plt.show()
```

```python
# BuildingAge vs Temperature

plt.figure(figsize=(8, 5))
sns.lineplot(x='BuildingAge', y='Temperature', data=energy_data,color='green')
plt.title('Building Age vs. Temperature')
plt.xlabel('Building Age (years)')
plt.ylabel('Temperature (C)')
plt.show()
```

```python
# BuildingAge vs EnergyConsumption

plt.figure(figsize=(8, 5))
sns.lineplot(x='BuildingAge', y='EnergyConsumption', data=energy_data, color='purple')
plt.title('Building Age vs. Energy Consumption')
plt.xlabel('Building Age (years)')
plt.ylabel('Energy Consumption (kWh)')
plt.show()
```

```python
# Temperature vs EnergyConsumption

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Temperature', y='EnergyConsumption', data=energy_data, alpha=0.6, color='red')
plt.title('Temperature vs. Energy Consumption')
plt.xlabel('Temperature (C)')
plt.ylabel('Energy Consumption (kWh)')
plt.show()
```

# Summary

This report analyzes how buildings consume energy and what factors influence their energy use. It looks at building size, insulation, renewable energy, humidity, temperature, and age to find patterns. The goal is to suggest ways to reduce energy costs and improve energy efficiency.

# Methodology

a) The energy consumption in buildings was cleaned by checking for null, there were null values present in the renewable energy source and the null values were filled with 'None' and there were no duplicates, dropped unecessary columns and converted categorical values into numerical values using label encoding so it could be analyzed properly.

b) Calculated Mean, Median and Standard Deviation Using Descriptive Statistics.

c) To find the co-relation between the attributes used heatmap.

d) Performed Exploratory Data Analysis where visualizations were generated to analyze study the connection between energy consumption, renewable energy source, insulation types, temperature, sixe and age of the buildings 


# Key Findings

## Energy Consumption vs. Building Size

Larger buildings tend to consume more energy.

## Humidity vs. Insulation Type 

Buildings with higher humidity tend to have poorer insulation, potentially leading to higher energy usage for temperature control.

## Building Size vs. Insulation Type

Larger buildings tend to invest in better insulation, reducing overall energy waste.

##  Building Age vs. Temperature

Older buildings generally experience more extreme temperature variations.

##  Building Age vs. Energy Consumption

Older buildings tend to consume more energy, possibly due to outdated insulation and HVAC systems.

##  Temperature vs. Energy Consumption

Higher external temperatures lead to higher energy consumption, mainly due to increased cooling needs.
